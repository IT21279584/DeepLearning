```python
import matplotlib.pyplot as plt
exp_data = np.random.exponential(scale = 1.0, size = 10000)
uniform_data = np.random.rand(100000)
normal_data = np.random.rand(100000)
plt.hist(exp_data, bins = 100)
plt.hist(uniform_data, bins = 1000)
plt.hist(normal_data, bins = 100)
plt.show()
```

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue",
label="Expotential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="red",
label="Uniform")
plt.hist(normal_data, density=True, bins=1, histtype="step", color="green",
label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid of values for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plotting the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Setting labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('$Z = X^2 + Y^2$')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
pokemon = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin1')

print(pokemon.columns)

features = pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Calculate the Pearson and Spearman correlation coefficient matrices
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

# Plot the Pearson correlation heatmap
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation')

# Plot the Spearman correlation heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
